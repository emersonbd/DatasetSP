```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Filtrar para RJ (cópia explícita)
dados_rj2022 = dados2022[dados2022['uf'] == 'RJ'].copy()
dados_rj2023 = dados2023[dados2023['uf'] == 'RJ'].copy()

# Obter lista de eventos únicos
eventos_unicos = sorted(dados_rj2023['evento'].unique())

# Dropdown para selecionar o evento
dropdown = widgets.Dropdown(
    options=eventos_unicos,
    value=eventos_unicos[0],  # Define o valor inicial como o primeiro evento na lista
    description='Evento:',
)

# Função para preencher meses ausentes
def preencher_meses(dados_agrupados):
    todos_os_meses = pd.DataFrame({'MES': range(1, 13)})
    dados_completos = pd.merge(todos_os_meses, dados_agrupados, on='MES', how='left')
    dados_completos['total_eventos'] = dados_completos['total_eventos'].fillna(0)  # Substituir diretamente
    return dados_completos

# Função para gerar o gráfico
def plot_evento_por_mes_e_ano(evento):
    # Filtrar os dados pelo evento selecionado
    dados_evento2022 = dados_rj2022[dados_rj2022['evento'] == evento].copy()
    dados_evento2023 = dados_rj2023[dados_rj2023['evento'] == evento].copy()
    
    # Criar uma coluna auxiliar para priorizar os valores
    for dados in [dados_evento2022, dados_evento2023]:
        dados['total_calculado'] = dados['total'].fillna(0) + dados['total_vitima'].fillna(0)
    
    # Agrupar por mês e somar o total calculado
    dados_agrupados2022 = dados_evento2022.groupby('MES')['total_calculado'].sum().reset_index(name='total_eventos')
    dados_agrupados2023 = dados_evento2023.groupby('MES')['total_calculado'].sum().reset_index(name='total_eventos')
    
    # Preencher meses ausentes
    dados_agrupados2022 = preencher_meses(dados_agrupados2022)
    dados_agrupados2023 = preencher_meses(dados_agrupados2023)
    
    # Calcular os totais de eventos por ano
    total_eventos_2022 = dados_agrupados2022['total_eventos'].sum()
    total_eventos_2023 = dados_agrupados2023['total_eventos'].sum()

    # Criar o gráfico
    plt.figure(figsize=(14, 7))
    
    # Plotar uma linha para cada ano com totais na legenda
    plt.plot(dados_agrupados2022['MES'], dados_agrupados2022['total_eventos'], marker='o', label=f'Ano 2022 - Total: {total_eventos_2022:,.0f}', color='blue')
    plt.plot(dados_agrupados2023['MES'], dados_agrupados2023['total_eventos'], marker='o', label=f'Ano 2023 - Total: {total_eventos_2023:,.0f}', color='red')
    
    # Adicionando as anotações para os valores mensais
    for mes, total_vitima in zip(dados_agrupados2022['MES'], dados_agrupados2022['total_eventos']):
        plt.text(mes, total_vitima + 5, f'{total_vitima:,.0f}', color='blue', ha='center', fontsize=10, fontweight='bold',
                 bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.3'))
    
    for mes, total_vitima in zip(dados_agrupados2023['MES'], dados_agrupados2023['total_eventos']):
        plt.text(mes, total_vitima + 5, f'{total_vitima:,.0f}', color='red', ha='center', fontsize=10, fontweight='bold',
                 bbox=dict(facecolor='white', edgecolor='red', boxstyle='round,pad=0.3'))

    # Configurações do gráfico
    plt.title(f'Total de Eventos "{evento}" por Mês no RJ (Por Ano)', fontsize=16)
    plt.xlabel('Mês', fontsize=14)
    plt.ylabel('Total de Eventos', fontsize=14)
    plt.xticks(ticks=range(1, 13), labels=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'], rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(title='Ano', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    
    # Exibir o gráfico
    plt.show()

# Exibir o widget interativo
output = widgets.interactive_output(plot_evento_por_mes_e_ano, {'evento': dropdown})
display(dropdown, output)
```

```python
# Carregar a planilha do Excel
arquivo_excel = "C:/Users/Emers/OneDrive/Workspace EBD/Doutorado/Disciplinas/Segundo Período/Dados SINESP/indicadoressegurancapublicamunic.xlsx"

# Função para ler todas as abas de uma planilha e sumarizar os dados
def ler_e_sumarizar_planilha(arquivo_excel):
    # Ler todas as abas da planilha
    todas_abas = pd.read_excel(arquivo_excel, sheet_name=None)

    # Inicializar um DataFrame vazio para armazenar os dados sumarizados
    df_summarizado = pd.DataFrame()

    # Iterar sobre cada aba e concatenar os dados
    for nome_aba, df3 in todas_abas.items():
        df3['Aba'] = nome_aba  # Adicionar uma coluna para identificar a aba
        df_summarizado = pd.concat([df_summarizado, df3], ignore_index=True)

    # Exibir o DataFrame sumarizado
    print(df_summarizado)

    # Retornar o DataFrame sumarizado
    return df_summarizado

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

def ler_planilha(arquivo_excel_munic):
    """
    Lê todas as abas de uma planilha Excel e concatena os dados em um único DataFrame.

    Parâmetros:
    arquivo_excel_munic (str): Caminho para o arquivo Excel.

    Retorna:
    pd.DataFrame: DataFrame concatenado com dados de todas as abas.
    """
    try:
        todas_abas = pd.read_excel(arquivo_excel_munic, sheet_name=None)
        df_concatenado = pd.DataFrame()

        for nome_aba, dfFull in todas_abas.items():
            dfFull['Aba'] = nome_aba  # Adicionar uma coluna para identificar a aba
            df_concatenado = pd.concat([df_concatenado, dfFull], ignore_index=True)

        return df_concatenado
    except Exception as e:
        print(f"Erro ao ler o arquivo Excel: {e}")
        return pd.DataFrame()

def obter_municipios_por_estado(df_concatenado):
    """
    Obtém uma lista de municípios agrupados por estado.

    Parâmetros:
    df_concatenado (pd.DataFrame): DataFrame com os dados concatenados.

    Retorna:
    dict: Dicionário com estados como chaves e listas de municípios como valores.
    """
    return df_concatenado.groupby('Sigla UF')['Município'].apply(list).to_dict()

import matplotlib.dates as mdates

def plot_vitimas(df_concatenado, estado, municipio):
    """
    Plota um gráfico aprimorado do somatório de vítimas de crime por mês/ano para um município específico,
    com o total de vítimas exibido na legenda e melhorias visuais.

    Parâmetros:
    df_concatenado (pd.DataFrame): DataFrame com os dados concatenados.
    estado (str): Sigla do estado.
    municipio (str): Nome do município.
    """
    df_filtered = df_concatenado[(df_concatenado['Sigla UF'] == estado) & (df_concatenado['Município'] == municipio)]

    # Converter 'Mês/Ano' para datetime, se necessário
    if not pd.api.types.is_datetime64_any_dtype(df_filtered['Mês/Ano']):
        df_filtered['Mês/Ano'] = pd.to_datetime(df_filtered['Mês/Ano'], errors='coerce')

    # Remover entradas com datas inválidas
    df_filtered = df_filtered.dropna(subset=['Mês/Ano'])

    # Agrupar por 'Mês/Ano' e somar as vítimas
    df_grouped = df_filtered.groupby(['Mês/Ano']).sum(numeric_only=True).reset_index()

    # Verificar se há dados para plotar
    if 'Vítimas' not in df_grouped or df_grouped.empty:
        print(f"Sem dados disponíveis para {municipio}, {estado}.")
        return

    # Calcular o total de vítimas
    total_vitimas = df_grouped['Vítimas'].sum()

    # Criar a legenda com o total de vítimas
    legenda = f'{municipio} - {estado} (Total: {total_vitimas:,.0f})'

    # Melhorias visuais no gráfico
    plt.figure(figsize=(14, 8))
    
    # Adicionar uma linha suave e pontos marcadores
    plt.plot(df_grouped['Mês/Ano'], df_grouped['Vítimas'], marker='o', color='tab:blue', linestyle='-', linewidth=2, markersize=6, label=legenda)
    
    # Adicionar título mais descritivo
    plt.title(f'Somatório de Vítimas de Crime no município {municipio}, {estado} ao Longo do Tempo\n(Total de Vítimas: {total_vitimas:,.0f})', fontsize=16)
    
    # Ajustar as labels dos eixos
    plt.xlabel('Mês/Ano', fontsize=14)
    plt.ylabel('Número de Vítimas', fontsize=14)

    # Formatando as labels de data para que fiquem legíveis
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b/%Y'))  # Formato de data (Ex: Jan/2020)

    # Adicionar uma grid mais suave para ajudar na visualização
    plt.grid(True, linestyle='--', alpha=0.7)

    # Ajustar a legenda para melhor visualização
    plt.legend(title='Legenda', loc='upper left', fontsize=12, title_fontsize=14)

    # Adicionar anotações para destacar o total de vítimas
    plt.annotate(f'Total: {total_vitimas:,.0f}', xy=(0.95, 0.05), xycoords='axes fraction', 
                 horizontalalignment='right', verticalalignment='bottom', fontsize=12, 
                 bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray', boxstyle='round,pad=0.5'))

    # Ajustar layout para evitar sobreposição
    plt.tight_layout()
    
    # Exibir o gráfico
    plt.show()


arquivo_excel_munic = 'C:/Users/Emers/OneDrive/Workspace EBD/Doutorado/Disciplinas/Segundo Período/Dados SINESP/indicadoressegurancapublicamunic.xlsx'
df_concatenado = ler_planilha(arquivo_excel_munic)

# Obter as listas de estados e municípios únicos
estados = df_concatenado['Sigla UF'].unique()
municipios_por_estado = obter_municipios_por_estado(df_concatenado)

# Widgets interativos
estado_widget = widgets.Dropdown(options=estados, description='Estado:')
municipio_widget = widgets.Dropdown(description='Município:')

# Função para atualizar os municípios com base no estado selecionado
def atualizar_municipios(change):
    estado = change['new']
    municipios = municipios_por_estado.get(estado, [])
    municipio_widget.options = municipios

# Observar mudanças no widget de estado
estado_widget.observe(atualizar_municipios, names='value')

# Atualizar municípios inicialmente
atualizar_municipios({'new': estados[0] if estados.size > 0 else None})

# Plotar gráfico ao selecionar estado e município
interact(lambda estado, municipio: plot_vitimas(df_concatenado, estado, municipio),
         estado=estado_widget, municipio=municipio_widget)
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import geopandas as gpd
from matplotlib import cm

# Mapeamento de estados para regiões
mapa_regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste', 
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste', 
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste', 
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste', 
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte', 
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}

# Carregar o arquivo GeoJSON dos estados do Brasil
geojson_path = 'C:/Users/Emers/OneDrive/Workspace EBD/Doutorado/Disciplinas/Segundo Período/brazil-states.geojson'
gdf = gpd.read_file(geojson_path)

# Adicionar a coluna 'Região' ao GeoDataFrame
gdf['Região'] = gdf['sigla'].map(mapa_regioes)

# 1. Função para plotar o mapa com as vítimas consolidadas por estado e ano
def plotar_mapa_com_vitimas(regiao, ano, esquema_cores):
    # Se a opção "Brasil" for selecionada, usar todos os estados
    if regiao == 'Brasil':
        gdf_regiao = gdf  # Não filtra por região, mostra todo o Brasil
        df_regiao = df_summarizado[df_summarizado['ANO'] == ano]  # Usamos todos os dados para o ano selecionado
    else:
        # Filtrar o GeoDataFrame para a região selecionada
        gdf_regiao = gdf[gdf['Região'] == regiao]
        df_regiao = df_summarizado[(df_summarizado['Região'] == regiao) & (df_summarizado['ANO'] == ano)]
    
    # Consolidar o número de vítimas por estado na região ou Brasil para o ano selecionado
    vitimas_por_estado = df_regiao.groupby('UF')['Vítimas'].sum()

    # Juntar os dados de vítimas ao GeoDataFrame com base no estado
    gdf_regiao = gdf_regiao.set_index('sigla').join(vitimas_por_estado).reset_index()

    # Aumentar o tamanho do mapa
    fig, ax = plt.subplots(figsize=(28, 22))  # Tamanho maior (30x24)

    # Plotar o mapa com o esquema de cores escolhido
    gdf_regiao.plot(ax=ax, column='Vítimas', cmap=esquema_cores, legend=True, 
                    legend_kwds={'label': "Número de Vítimas", 'orientation': "horizontal"})

    # Adicionar título
    ax.set_title(f'Mapa das Vítimas por Estado - Ano: {ano} - Região: {regiao}' if regiao != 'Brasil' else f'Mapa das Vítimas por Estado - Ano: {ano} - Brasil', fontsize=18, fontweight='bold')
    
    # Adicionar rótulos com o número de vítimas sobre os estados
    for idx, row in gdf_regiao.iterrows():
        # Verifica se o número de vítimas é maior que um valor mínimo para mostrar o rótulo
        if row['Vítimas'] > 0:  # Pode ajustar o valor mínimo, ex: 10
            # Posição do rótulo (centroide do estado)
            x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
            
            # Ajuste dinâmico da posição para evitar sobreposição (deslocamento com base no centroide)
            ax.annotate(f"{int(row['Vítimas']):,}", 
                        xy=(x, y), 
                        xytext=(6, 6),  # Deslocamento do texto (evita sobreposição)
                        textcoords="offset points", 
                        ha='center', 
                        fontsize=16, 
                        color='black', 
                        fontweight='bold', 
                        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))  # Caixa de fundo para destacar o texto

    # Adicionar grid para o gráfico (melhorando a visibilidade)
    ax.grid(True, linestyle='--', alpha=0.7)

    # Ajustar a visibilidade do grid e melhorar o layout
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('')
    ax.set_ylabel('')

    # Mostrar o mapa
    plt.show()

# 2. Criar o widget para selecionar a região ou Brasil
dropdown_regioes = widgets.Dropdown(
    options=['Brasil'] + list(gdf['Região'].unique()),  # Incluindo a opção 'Brasil'
    description='Região:',
    disabled=False
)

# 3. Criar o widget para escolher o ano
dropdown_ano = widgets.Dropdown(
    options=sorted(df_summarizado['ANO'].unique()),  # Todos os anos disponíveis
    description='Ano:',
    disabled=False
)

# 4. Criar o widget para escolher o esquema de cores
dropdown_cores = widgets.Dropdown(
    options=['Blues', 'YlOrRd', 'viridis', 'plasma', 'coolwarm', 'cividis'],
    description='Esquema de Cores:',
    disabled=False
)

# Função interativa que conecta a seleção de região, ano e esquema de cores ao gráfico
widgets.interactive(plotar_mapa_com_vitimas, regiao=dropdown_regioes, ano=dropdown_ano, esquema_cores=dropdown_cores)
```
